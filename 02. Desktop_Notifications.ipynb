```python
import time

from plyer import notification

if __name__ == "__main__":
    while True:
        notification.notify( title = "Alert messege...!",
                           message = "Break time>>>",
                           timeout = 10)
        time.sleep(60)
```

# It will Run and continuously receive notifications every 60 sec or until the time you set due to the while loop defined in the code
