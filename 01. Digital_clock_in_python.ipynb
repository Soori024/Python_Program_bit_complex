```python
''' A Digital clock using python. This is a simple task to get started with the Tkinter library in Python, which is a built-in package that comes with Python. 
Tkinter has some cool features that can be used to build simple apps.'''

from tkinter import Label, Tk
import time

window = Tk()
window.title("Digital Clock View")
window.geometry("420x150")
window.resizable(1,1)

font = ("Boulder", 72, 'bold')
background = "#00FF00"
foreground = "#363529"
width_border = 28

label = Label(window, font=font, bg=background, fg=foreground, bd=width_border) 
label.grid(row=0, column=1)

def digital_clock(): 
    time_live = time.strftime("%H:%M:%S")
    label.config(text=time_live) 
    label.after(200, digital_clock)

digital_clock()
window.mainloop()
```
